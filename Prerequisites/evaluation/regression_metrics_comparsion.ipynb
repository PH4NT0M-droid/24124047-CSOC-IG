## Evaluation Metrics Comparison for above 3 parts

```python
import matplotlib.pyplot as plt
import numpy as np

methods = ['Pure Python', 'NumPy', 'Scikit-learn']

train_mae = [49499.41, 49499.40, 48352.88]
test_mae = [48974.65, 48974.66, 50277.55]

train_rmse = [67716.47, 67716.47, 66319.52]
test_rmse = [68926.54, 68926.54, 69407.05]

train_r2 = [0.6442, 0.6442, 0.6709]
test_r2 = [0.6739, 0.6739, 0.6324]

train_time = [80.13, 0.63, 0.0070]

fig, axs = plt.subplots(2, 2, figsize=(14, 10))
bar_width = 0.4
x = np.arange(len(methods))

# 1. MAE Plot
axs[0, 0].bar(x - bar_width/2, train_mae, width=bar_width, label='Train MAE', color='skyblue')
axs[0, 0].bar(x + bar_width/2, test_mae, width=bar_width, label='Test MAE', color='dodgerblue')
axs[0, 0].set_title('Mean Absolute Error (MAE)')
axs[0, 0].set_xticks(x)
axs[0, 0].set_xticklabels(methods)
axs[0, 0].legend()

# 2. RMSE Plot
axs[0, 1].bar(x - bar_width/2, train_rmse, width=bar_width, label='Train RMSE', color='lightcoral')
axs[0, 1].bar(x + bar_width/2, test_rmse, width=bar_width, label='Test RMSE', color='firebrick')
axs[0, 1].set_title('Root Mean Squared Error (RMSE)')
axs[0, 1].set_xticks(x)
axs[0, 1].set_xticklabels(methods)
axs[0, 1].legend()

# 3. R2 Score Plot
axs[1, 0].bar(x - bar_width/2, train_r2, width=bar_width, label='Train R', color='lightgreen')
axs[1, 0].bar(x + bar_width/2, test_r2, width=bar_width, label='Test R', color='seagreen')
axs[1, 0].set_title('R Score')
axs[1, 0].set_xticks(x)
axs[1, 0].set_xticklabels(methods)
axs[1, 0].legend()

# 4. Training Time Plot
axs[1, 1].bar(x, train_time, width=bar_width, color='mediumpurple')
axs[1, 1].set_title('Training Time (s)')
axs[1, 1].set_xticks(x)
axs[1, 1].set_xticklabels(methods)

plt.suptitle('Regression Metric Comparison Across Implementations', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
